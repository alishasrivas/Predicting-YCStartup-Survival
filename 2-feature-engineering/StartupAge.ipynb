```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re

for company in scraped_og['Company']:
    # URL of the website to scrape
    url = f"https://www.ycombinator.com/companies/{company}"

    # Send a GET request to the URL
    response = requests.get(url)

    # Parse the page content with BeautifulSoup
    soup = BeautifulSoup(response.content, "html.parser")

    # Find all div elements with the specified class
    divs = soup.find_all("div", class_="ycdc-card space-y-1.5 sm:w-[300px]")

    yc_url = f"https://www.ycombinator.com/companies/{company}"
    yc_response = requests.get(yc_url)
    soup = BeautifulSoup(yc_response.content, "html.parser")
    result = soup.find('a', href=lambda x: x and x.startswith("/companies?batch="))
    try:
        href_value = result['href']
    except Exception as e:
        pass
    batch_value = int(re.findall(r'\d+', str(href_value.split("batch=")[-1]))[0]) + 2000

    # Extract and print the text from each div
    for div in divs:
        year_founded = int(re.findall(r'\d+', div.get_text(strip=True))[0])

    flag = None
    if batch_value < year_founded:
        flag = 'Negative Value'
    elif batch_value - year_founded > 10:
        flag = 'Unusually High Value'

    data = [{'slug': company, 'year_founded': year_founded, 'yc_batch': batch_value, 'age': batch_value - year_founded, 'flag': flag}]
    startup_age = pd.concat([startup_age, pd.DataFrame(data)], ignore_index=True)

    print(f"{company} YC Graduation Date: {batch_value}")
    print(f"{company} Founded Date: {year_founded}")
    print(f"{company} Startup Age: {batch_value - year_founded}")

```
